```python
from googleplaces import GooglePlaces, types, lang
import pandas as pd
import time
```

```python
API_KEY = ''
google_places = GooglePlaces(API_KEY)
```

```python
query_result_brk = google_places.nearby_search(
        location='Brooklyn', types=[types.TYPE_RESTAURANT] or [types.TYPE_FOOD], keyword='Restaurant')

query_result_man = google_places.nearby_search(
        location='Manhattan', types=[types.TYPE_RESTAURANT] or [types.TYPE_FOOD], keyword='Restaurant')

query_result_brx = google_places.nearby_search(
        location='Bronx', types=[types.TYPE_RESTAURANT] or [types.TYPE_FOOD], keyword='Restaurant')

query_result_que = google_places.nearby_search(
        location='Queens', types=[types.TYPE_RESTAURANT] or [types.TYPE_FOOD], keyword='Restaurant')

query_result_sta = google_places.nearby_search(
        location='Staten Island', types=[types.TYPE_RESTAURANT] or [types.TYPE_FOOD], keyword='Restaurant')
```

```python
placelist = []
```

```python
for i in range(3):
    for place in query_result_brk.places:
        place.get_details()
        data = {
            'Name': place.name,
            'Location': place.geo_location,
            'Rating': place.rating,
            #'Price': place.details['price_level'],
            'Borough': 'Brooklyn'
        }
        placelist.append(data)
      
    if query_result_brk.has_next_page_token:
        query_result_next_page = google_places.nearby_search(pagetoken = query_result_brk.next_page_token)
    
    query_result_brk = query_result_next_page
    
    #print(i)
    #time.sleep(2)
```
